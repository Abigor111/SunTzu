```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'Tempo': pd.date_range('2024-01-01', periods=100),
    'Vendas': np.random.randint(0, 100, size=100),
    'Lucro': np.random.uniform(0, 1000, size=100),
    'Categoria': np.random.choice(['A', 'B', 'C'], size=100),
    'Idade': np.random.randint(18, 70, size=100)
}

df = pd.DataFrame(data)

# Visualizando as primeiras linhas do DataFrame
df.head()
```

```python
def scatter_plot(self, x, y, title="", xlabel="", ylabel="", grid=False):
    if xlabel is None:
        xlabel = str(x)
    if ylabel is None:
        ylabel = str(y)
    self.plot.scatter(x=x, y=y, title=title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(grid)
    plt.legend()
    plt.show()
```

```python
def line_plot(self, x, y, title="", xlabel="", ylabel="", grid=False):
    if xlabel is None:
        xlabel = str(x)
    if ylabel is None:
        ylabel = str(y)
    print(title is None)
    self.plot.line(x=x, y=y, title=title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(grid)
    plt.legend()
    plt.show()
```

```python
line_plot(df, 'Tempo', 'Vendas', title="teste", xlabel="Sells", grid=True)
```
